# Train data 만들기

```python
visit=pd.read_csv('강원 방문자.csv', encoding='cp949')
con=pd.read_csv('강원 관광소비.csv', encoding='cp949')
work=pd.read_csv('강원 동반.csv', encoding='cp949')
pur=pd.read_csv('강원 목적.csv', encoding='cp949')


visit=visit[['기준연월', '방문자 수']]


con = con.pivot(index='기준연월', columns='업종대분류명', values='소비액(천원)').reset_index()
con = con.rename(columns={'전체': '전체소비'})

work = work.pivot(index='기준연월', columns='동반유형명', values='언급건수').reset_index()
work['합계'] = work.sum(axis=1)
work['워케이션 비율']=work['회사동료']/work['합계']
work=work[['기준연월', '워케이션 비율']]

work=pd.merge(work, visit, on='기준연월')
work['워케이션']=round(work['워케이션 비율']*work['방문자 수'])
work=work[['기준연월', '워케이션']]

pur = pur.pivot(index='기준연월', columns='목적지 유형', values='목적지 검색량').reset_index()
pur['합계'] = pur.sum(axis=1)

for i in range(18):
    for j in range(1, 10):
        pur.iloc[i, j] = round(pur.iloc[i, j] / pur.iloc[i, 10], 3)
        
pur = pur.drop(columns='합계')
pur=pd.merge(pur, work, on='기준연월')

for i in range(18):
    for j in range(1, 10):
        pur.iloc[i, j] = round(pur.iloc[i, j] * pur.iloc[i, 10])
        
pur = pur.drop(columns='워케이션')

df = pd.merge(pd.merge(pd.merge(visit, work, on='기준연월'), pur, on='기준연월'), con, on='기준연월')
for i in range(18):
    df.iloc[i, 0]='강원'+str(df.iloc[i, 0])
df
```

```python
def makedf (df, 지역):
    a=지역+' 방문자.csv'
    b=지역+' 관광소비.csv'
    c=지역+' 동반.csv'
    d=지역+' 목적.csv'
    
    visit=pd.read_csv(a, encoding='cp949')
    con=pd.read_csv(b, encoding='cp949')
    work=pd.read_csv(c, encoding='cp949')
    pur=pd.read_csv(d, encoding='cp949')
    
    visit=visit[['기준연월', '방문자 수']]

    con = con.pivot(index='기준연월', columns='업종대분류명', values='소비액(천원)').reset_index()
    con = con.rename(columns={'전체': '전체소비'})

    work = work.pivot(index='기준연월', columns='동반유형명', values='언급건수').reset_index()
    work['합계'] = work.sum(axis=1)
    work['워케이션 비율']=work['회사동료']/work['합계']
    work=work[['기준연월', '워케이션 비율']]

    work=pd.merge(work, visit, on='기준연월')
    work['워케이션']=round(work['워케이션 비율']*work['방문자 수'])
    work=work[['기준연월', '워케이션']]

    pur = pur.pivot(index='기준연월', columns='목적지 유형', values='목적지 검색량').reset_index()
    pur['합계'] = pur.sum(axis=1)

    for i in range(18):
        for j in range(1, 10):
            pur.iloc[i, j] = round(pur.iloc[i, j] / pur.iloc[i, 10], 3)
        
    pur = pur.drop(columns='합계')
    pur=pd.merge(pur, work, on='기준연월')

    for i in range(18):
        for j in range(1, 10):
            pur.iloc[i, j] = round(pur.iloc[i, j] * pur.iloc[i, 10])
        
    pur = pur.drop(columns='워케이션')

    df1 = pd.merge(pd.merge(pd.merge(visit, work, on='기준연월'), pur, on='기준연월'), con, on='기준연월')
    for i in range(18):
        df1.iloc[i, 0]=지역+str(df1.iloc[i, 0])

    df = pd.concat([df, df1], ignore_index=True)
    
    
    return df
```

# 모델링 실험

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 입력 변수 x와 출력 변수 y로 나누기
x_train = df[['기타관광', '레저스포츠', '문화관광', '쇼핑', '숙박', '역사관광', '음식', '자연관광', '체험관광']]
y_train = df[['쇼핑업', '숙박업', '식음료업', '여가서비스업', '운송업', '전체소비']]

# 데이터프레임을 파이토치 텐서로 변환
x_train_tensor = torch.tensor(x_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)

# 모델 구축
class MultivariateModel(nn.Module):
    def __init__(self):
        super(MultivariateModel, self).__init__()
        self.fc1 = nn.Linear(9, 10)  # 첫 번째 은닉층, 9개의 입력 노드와 10개의 뉴런
        self.fc2 = nn.Linear(10, 5)  # 두 번째 은닉층, 10개의 뉴런과 5개의 뉴런
        self.fc3 = nn.Linear(5, 6)   # 출력층, 5개의 뉴런과 6개의 출력 노드

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # 첫 번째 은닉층의 활성화 함수는 ReLU
        x = torch.relu(self.fc2(x))  # 두 번째 은닉층의 활성화 함수는 ReLU
        x = self.fc3(x)  # 출력층에는 활성화 함수를 사용하지 않음
        return x

model = MultivariateModel()

# 손실 함수와 옵티마이저 정의
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 모델 학습
for epoch in range(100):
    optimizer.zero_grad()
    outputs = model(x_train_tensor)  # 수정: x_train_tensor 사용
    loss = criterion(outputs, y_train_tensor)  # 수정: y_train_tensor 사용
    loss.backward()
    optimizer.step()

# 새로운 데이터로 예측
new_data = torch.tensor([[10, 20, 30, 40, 50, 60, 70, 80, 90]], dtype=torch.float32)  # 수정: dtype 지정
predictions = model(new_data)

print("Predictions:", predictions)
```
