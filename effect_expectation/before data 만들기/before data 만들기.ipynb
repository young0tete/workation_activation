```python
def makedf (df, 지역):
    a=지역+' 목적.csv'
    
    pur=pd.read_csv(a, encoding='cp949')

    pur = pur.pivot(index='기초지자체', columns='카테고리중분류명 ', values='중분류 검색건수').reset_index()
    pur = pur.fillna(0)
    pur['합계'] = pur.sum(axis=1)

    for index, row in pur.iterrows():
        for j in range(1, 10):
            pur.iloc[index, j] = round(pur.iloc[index, j] / pur.iloc[index, 10], 5)
        
    pur = pur.drop(columns='합계')

    for index, row in pur.iterrows():
        pur.iloc[index, 0]=지역+' '+str(pur.iloc[index, 0])

    df = pd.concat([df, pur], ignore_index=True)
    
    return df
```
