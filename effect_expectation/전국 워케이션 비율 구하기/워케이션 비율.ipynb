```python
import pandas as pd
import numpy as np
```

```python
visit=pd.read_csv('강원 방문자.csv', encoding='cp949')
con=pd.read_csv('강원 관광소비.csv', encoding='cp949')
work=pd.read_csv('강원 동반.csv', encoding='cp949')
pur=pd.read_csv('강원 목적.csv', encoding='cp949')


visit=visit[['기준연월', '방문자 수']]

work = work.pivot(index='기준연월', columns='동반유형명', values='언급건수').reset_index()
work['합계'] = work.sum(axis=1)
work['워케이션 비율']=work['회사동료']/work['합계']
work=work[['기준연월', '워케이션 비율']]


df0 = pd.merge(visit, work, on='기준연월')

data = {
    '지역': ['강원'],
    '워케이션 비율': [round(df0['워케이션 비율'].mean(), 5)]
}

df = pd.DataFrame(data)
df
```

```python
def makedf (df, 지역):
    a=지역+' 방문자.csv'
    b=지역+' 관광소비.csv'
    c=지역+' 동반.csv'
    d=지역+' 목적.csv'
    
    visit=pd.read_csv(a, encoding='cp949')
    con=pd.read_csv(b, encoding='cp949')
    work=pd.read_csv(c, encoding='cp949')
    pur=pd.read_csv(d, encoding='cp949')
    
    visit=visit[['기준연월', '방문자 수']]

    work = work.pivot(index='기준연월', columns='동반유형명', values='언급건수').reset_index()
    work['합계'] = work.sum(axis=1)
    work['워케이션 비율']=work['회사동료']/work['합계']
    work=work[['기준연월', '워케이션 비율']]


    df0 = pd.merge(visit, work, on='기준연월')

    data = {
        '지역': [지역],
        '워케이션 비율': [round(df0['워케이션 비율'].mean(), 5)]
    }

    df1 = pd.DataFrame(data)


    df = pd.concat([df, df1], ignore_index=True)
    
    return df
```
