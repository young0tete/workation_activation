```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('raw data.csv')
df
```

```python
df.loc['합계'] = df.iloc[:, 1:].sum()
df.iat[69, 0] = '합계'
df['합계'] = df.sum(axis=1)
df
```

```python
#카테고리별 해당 지역의 비율
df1=df.copy()

for i in range(1, 10):
    for j in range(69):
        df1.iat[j, i] = df1.iat[j,i]/df1.iat[69,i]

df1
```

```python
#역내 해당 카테고리의 비율
df2=df.copy()

for i in range(69):
    for j in range(1, 10):
        df2.iat[i, j] = df2.iat[i, j]/df2.iat[i, 10]

df2
```

```python
#카테고리별 rescaling
df3=df.copy()
df3 = df3.drop(df3.index[-1])

for i in range(1, 10):
    min_val = df3.iloc[:, i].min() 
    max_val = df3.iloc[:, i].max() 
    
    for j in range(69):
        df3.iat[j, i] = (df3.iat[j, i] - min_val) / (max_val - min_val)

df3
```

```python
df4=df2+df3
df4.iloc[:, 0] = df2.iloc[:, 0]
df4 = df4.drop(df4.index[-1])
df4
```

```python
# 두 개의 시리즈 생성 (예시)
s1 = pd.Series(df4.iloc[:, 0], name='시군구')
s2 = pd.Series(df4.iloc[:, 1]+df4.iloc[:, 2], name='스포츠')
s3 = pd.Series(df4.iloc[:, 3]+df4.iloc[:, 4], name='역사')
s4 = pd.Series(df4.iloc[:, 5]+df4.iloc[:, 6], name='자연힐링')
s5 = pd.Series(df4.iloc[:, 7]+df4.iloc[:, 8], name='문화예술')
s6 = pd.Series(df4.iloc[:, 9]+df4.iloc[:, 9]+df4.iloc[:, 9], name='맛집')

# 데이터프레임으로 합치기 (열 방향으로 합침)
df5 = pd.concat([s1, s2, s3, s4, s5, s6], axis=1)
df5
```

```python
df5.to_csv('전처리 ver4.csv', index=False)
```
