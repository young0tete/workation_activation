```python
import pandas as pd
```

```python
df = pd.read_csv('전처리 ver5.csv')
```

```python
df
```

```python
melted_df = df.melt(id_vars='시군구',value_vars=['스포츠','역사','자연힐링','문화예술','맛집'], var_name = '피벗 필드명', value_name = '피벗 필드값')
```

```python
melted_df = melted_df.rename(columns={'시군구' : '지역'})
```

```python
melted_df.to_csv('피벗_ver5.csv', index=False)
```
